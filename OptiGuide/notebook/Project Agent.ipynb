````python
import autogen
from optiguide.experimental.explorer import ProjectAgent
from autogen import UserProxyAgent
import os

config_list_gpt4 = autogen.config_list_from_json(
    "OAI_CONFIG_LIST",
    filter_dict={
        "model": ["gpt-4", "gpt-4-0314", "gpt4", "gpt-4-32k", "gpt-4-32k-0314", "gpt-4-32k-v0314"],
    },
)


config_list_gpt3 = autogen.config_list_from_json(
    "OAI_CONFIG_LIST",
    filter_dict={
        "model": ["gpt-35-turbo", "gpt-3.5-turbo"],
    },
)
````

## Example: Coffee Dataset

We use the [Coffee Dataset](https://github.com/BeibinLi/Coffee_Roasting_Dataset) as the example application, where users


```bash
git clone https://github.com/BeibinLi/Coffee_Roasting_Dataset ~/
```

````python
repo_loc = os.path.expanduser("~/Coffee_Roasting_Dataset/opti_coffee/")
assert os.path.exists(repo_loc)
````

````python
agent = ProjectAgent(
    name="Project-Agent",
     root=repo_loc,
    max_consecutive_auto_reply=30,
    llm_config={"config_list": config_list_gpt4, "seed": 42},
     description=""
)

user = autogen.UserProxyAgent(
    name="User_proxy",
    system_message="A human admin.",
    human_input_mode="NEVER",  # Try between ALWAYS or NEVER
    max_consecutive_auto_reply=0,
)
````

````python
# user.initiate_chat(agent, message="Plot the average of employee salary in a global map. Use the data and documents inside the repository.")
user.initiate_chat(agent, message="Read the datbase, and show me the average of employee salary per country.")
````
